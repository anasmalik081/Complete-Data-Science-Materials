```
# importing pandas
import pandas as pd
```

## **Pandas read_html**

Pandas has an inbuilt function,i.e., read_html which help us to read html file and convert tables into dataframe.

```
# reading an html file
html = pd.read_html("https://en.wikipedia.org/wiki/Mobile_country_code")
print(type(html))
```

```
# length of list
len(html)
```

```
# table content from html
html
```

```
# first table
html[0]
```

```
# second table
html[1].head()
```

```
# third table
html[2].head()
```

```
# reading a specific table from html file
html = pd.read_html("https://en.wikipedia.org/wiki/Economy_of_the_United_States", match="Government debt")
```

```
html[1].head()
```

```
# reading a specific table from html file
html = pd.read_html("https://en.wikipedia.org/wiki/Minnesota", match='Average daily maximum and minimum temperatures for selected cities in Minnesota')
```

```
html[0]
```

## **Pandas to_html**

Pandas has an inbuilt function,i.e., to_html which help us to write a dataframe to a html file.

```
# reading the same file
html = pd.read_html("https://en.wikipedia.org/wiki/Minnesota", match='Average daily maximum and minimum temperatures for selected cities in Minnesota')
html[0]
```

```
html[0].to_html('demo.html')
```
